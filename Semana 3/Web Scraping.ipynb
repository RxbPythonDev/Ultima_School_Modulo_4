# Web Scrapping

```python
import requests
from bs4 import BeautifulSoup
```

```python
url_base = 'https://lista.mercadolivre.com.br/'
```

```python
product_name = input('Escolha o produto: ')
```

```python
product_name
```

```python
nome_trat = product_name.replace(' ', '-')
```

```python
nome_trat
```

```python
#unindo a url e o produto
url_produto = url_base + nome_trat
```

```python
url_produto
```

```python
response = requests.get(url_produto)
print(response)
```

```python
site = BeautifulSoup(response.text, 'html.parser')
```

```python
dados = site.find_all("div", attrs= {"class" : "ui-search-result__content-wrapper shops__result-content-wrapper"})
```

```python
print(dados)
```

```python
produtos = []
precos = []

for x in dados:

  #nome do produto
  nome_mineiracao = x.find_all('h2')[0].text
  produtos.append(nome_mineiracao)

  #preço do produto
  preco_mineiracao = x.find_all('span', class_='price-tag ui-search-price__part shops__price-part')[0].text
  preco_trat = preco_mineiracao.split(' ')[0]
  precos.append(preco_trat)
```

```python
import pandas as pd
```

```python
dados_coletados = {
    'Nome': produtos,
    'Preco': precos
}
```

```python
tabela = pd.DataFrame(dados_coletados)
```
